```python
import os
import json
import numpy as np
```

```python
gameData = []
for root, dirs, files in os.walk('./data/fullGameRecord/'):
    cnt = 0
    for file in files:
        cnt += 1
        filename = './data/fullGameRecord/' + str(cnt) + '.json'
        with open(filename, 'r', encoding="utf8") as f:
            gameData.append(json.loads(f.read()))
```

```python
def dataClean(gameData, playerNum, observeNum):
    cnt = 0
    roleList = []
    frequency = np.zeros(200, dtype=int)

    for game in gameData:
        if game['numberOfPlayers'] == playerNum:
            cnt += 1
            if game['roles'] not in roleList:
                roleList.append(game['roles'])
                for i in range(len(roleList)):
                    if roleList[i] == game['roles']:
                        frequency[i] += 1
            else:
                for i in range(len(roleList)):
                    if roleList[i] == game['roles']:
                        frequency[i] += 1

    print('游戏人数为%d人的对局共%d局' %(playerNum, cnt))
    print('共出现%d种角色配置'%len(roleList))
    print('出现频率前%d名如下' %observeNum)

    x = frequency.argsort()
    for i in range(observeNum):
        pos = x[-i-1]
        print(frequency[pos])
        print(roleList[pos])
    print('\n')
```

```python
for i in range(5, 11):
    dataClean(gameData, i, 10)
```
